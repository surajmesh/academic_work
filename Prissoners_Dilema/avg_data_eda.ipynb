```python
plt.figure(figsize=(12, 5))
sns.distplot(data,bins=10)
```

```python
plt.figure(figsize=(15, 7))
plt.xlabel("Matches Played")
plt.ylabel("Average_score")
plt.title("Matches Played Vs Average score")
plt.legend("")
l=["abhishek","pankaj","akhi_sur","dn","ar","allD","ash","tft","allC","roshan","tft1","random1"]
for i in l:
    plt.plot(data["Matches Played"],data[i])
    plt.legend(l,fontsize=12)
```

```python
data1 =[data["abhishek"],data["dn"],data['ar'],data["akhi_sur"],data["allD"]]
fig = plt.figure(figsize =(10, 7))
ax = fig.add_axes([0, 0, 1, 1])

bp = ax.boxplot(data1)
plt.xticks([1, 2, 3,4,5], ['abhishek','dn','ar','akhi_sur','allD'])


ax.set_xlabel('Function name of Compitors')
ax.set_ylabel('Average Wining Scores')

```
