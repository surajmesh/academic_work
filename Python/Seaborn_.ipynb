# Seaborn Exercises

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
sns.jointplot(x='fare',y='age',data=titanic,height=7)
```

```python
sns.distplot(titanic['fare'],kde=False,bins=30,color='red')
```

```python
sns.boxplot(x='class',y='age',data=titanic,palette='rainbow',linewidth=3,dodge=False)
```

```python
sns.swarmplot(x='class',y='age',data=titanic,palette='Set1')
```

```python
sns.countplot(titanic['sex'],palette='Set3')
```

```python
import numpy as np
tc=titanic.corr()                                       #color
sns.heatmap(tc,annot=True,linewidth=2,linecolor='green',center=3)
plt.title('titanic.corr()')
```

```python
g = sns.FacetGrid(data=titanic,col='sex')
g.map(plt.hist,'age')
```
