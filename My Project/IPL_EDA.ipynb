```python
# Extracting the record where a team won batting first
battting_First = ipl[ipl['win_by_runs'] != 0]
battting_First.head()
```

```python
# Making  a histogram
plt.figure(figsize=(7,5))
plt.hist(battting_First['win_by_runs'],bins=10,color="green")
plt.title("Distribution Plot")
plt.xlabel("Run")
```

```python
# first batting whose win the match
battting_First['winner'].value_counts()
```

```python
# we drAW HIST for batting first winner Team
X_axis = list(battting_First['winner'].value_counts()[0:3].keys())
print(X_axis) # for catogorical
print("---------------------------------------------------------------")

Y_axis = list(battting_First['winner'].value_counts()[0:3])
print(Y_axis) # for numrical

plt.figure(figsize = (7,5))
plt.bar(X_axis,Y_axis,color =["Blue","yellow","red"])
```

```python
# Making a DoughtNut Chart for max win 3 team
plt.pie(Y_axis,labels=X_axis,autopct="%0.1f%%",shadow=10,radius=1.5)
plt.pie([1],colors="white")
```

```python
# for pie chart first batting  all winning team

plt.pie(battting_First['winner'].value_counts(),
        labels = battting_First['winner'].value_counts().keys(),radius = 2,autopct="%0.1f%%")
```

```python
# Extractiing those records where a team has after batting second !!
batting_second = ipl[ipl['win_by_wickets'] != 0 ]
batting_second.head()
```

```python
# plot hist for frequncey of w.r.t number of wicket
plt.hist(batting_second['win_by_wickets'],bins=20,color ="g")
```

```python
#Finding out the frequency of number of win w.r.t each time after batting second 
batting_second['winner'].value_counts()
```

```python
# Making a pie chart for Distribuition of most wins after batting second
plt.figure(figsize=(7,7))
plt.pie(batting_second['winner'].value_counts(),labels=batting_second['winner'].value_counts().keys(),autopct="%0.1f%%")
```
